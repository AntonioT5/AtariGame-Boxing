```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
scratch_ep = pd.read_csv("scratch_episodes.csv")
resnet_ep = pd.read_csv("resnet_episodes.csv")
scratch_wr = pd.read_csv("scratch_winrate.csv")
resnet_wir = pd.read_csv("resnet_winrate.csv")

scratch_ep.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(scratch_ep["episode"], scratch_ep["rolling_avg_30"], label="Scratch Double DQN", color="#003f5c", linewidth=2)
ax.plot(resnet_ep["episode"], resnet_ep["rolling_avg_30"], label="Pretrained ResNet", color="#008c54", linewidth=2)
ax.set_xlabel("Episode")
ax.set_ylabel("Reward (rolling average)")
ax.set_title("Reward Progress Over Training")
ax.legend()
ax.grid(alpha=0.3)
plt.savefig("reward_comparison.png", dpi=150)
plt.show()
```

```python
print("Scratch - mean reward: ", scratch_ep["rolling_avg_30"].mean(), " final: ", scratch_ep["rolling_avg_30"].iloc[-1])
print("ResNet - mean reward: ", resnet_ep["rolling_avg_30"].mean(), " final: ", resnet_ep["rolling_avg_30"].iloc[-1])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(scratch_wr["episode"], scratch_wr["win_rate"], marker="o", label="Scratch CNN", color="#2E86AB", linewidth=2)
ax.plot(resnet_wir["episode"], resnet_wir["win_rate"], marker="o", label="Pretrained ResNet", color="#E67E22", linewidth=2)
ax.set_xlabel("Episode")
ax.set_ylabel("Win Rate vs Random")
ax.set_title("Win Rate Over Training")
ax.set_ylim(-0.05, 1.05)
ax.legend()
ax.grid(alpha=0.3)
plt.savefig("winrate_comparison.png", dpi=150)
plt.show()
```

```python
print("Scratch - mean win rate: ", scratch_wr["win_rate"].mean(), " final: ", scratch_wr["win_rate"].iloc[-1])
print("ResNet - mean win rate: ", resnet_wir["win_rate"].mean(), " final: ", resnet_wir["win_rate"].iloc[-1])
```
